```python
#TASK 3 

import sklearn.linear_model
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error
#importing linear regression and its compononents

x_variable_list = ['fixed acidity', 'volatile acidity', 'citric acid', 'residual sugar',
       'chlorides', 'free sulfur dioxide', 'total sulfur dioxide', 'density',
       'pH', 'sulphates', 'alcohol']

# on x axis we will have all the features based on which we want to predict 
x = df[x_variable_list]

#on y axis we will have the feature we want to predict
y = df['quality']

# we us 80:20 in train test (as asked in task 3 )
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2)

print (x_train.shape, y_train.shape)
print (x_test.shape, y_test.shape)
```

```python
# The risk of sampling bias is low as our data set is large enough compared to number of attributes
# hands on machine learning book (chapter 1, page 53)
# i.e we dont need stratified sampling 
```

```python
print('Median of wine quality from training data set is:', y_train.median())
med = df['quality'].median()
y_train.describe()
```

```python
Length_y = len(y_train)
rmse = np.linalg.norm(y_train - med) / np.sqrt(Length_y)
print('The RMSE Value is:', rmse)
#task 3.a
```

LINEAR REGRESSION

```python
regular = LinearRegression()
regular.fit(x_train,y_train)
```

```python
train_pred = regular.predict(x_train)
print(train_pred)
```

```python
test_pred = regular.predict(x_test) 
print(test_pred)
```

```python
# predicting values for train and test data 
```

```python
# plotting a graph between predicted and true values
plt.scatter(y_train, train_pred, color = 'g',alpha=0.5)
plt.xlabel('True Values')
plt.ylabel('LM Predictions')
plt.title('Linear Model - LMP vs TV')

m = range(3,10,1)
n = range(3,10,1)
plt.plot(m,n, color = 'r', label='True = Predict', alpha = 0.5)
```

```python
# calculating rmse on our predicted values
train_rmse = mean_squared_error(train_pred, y_train) ** 0.5
print(train_rmse)
test_rmse = mean_squared_error(test_pred, y_test) ** 0.5
print(test_rmse)
```

```python
from sklearn import metrics
print('Mean Absolute Error:', metrics.mean_absolute_error(y_test, test_pred))
print('Mean Squared Error:', metrics.mean_squared_error(y_test, test_pred))
print('Root Mean Squared Error:', np.sqrt(metrics.mean_squared_error(y_test, test_pred)))
```

```python
accuracy = regular.score(x_test, y_test)
"Accuracy: {}%".format(int(round(accuracy * 100, 2)))
```

```python
from math import sqrt
lin_model_new = regular.predict(x_train)
lin_rmse = sqrt(mean_squared_error(y_train, lin_model_new))
print(lin_rmse)
```

K- NEAREST INTEGER 

```python
#Importing k-nearest neighbors Classifier model from sklearn
from sklearn.neighbors import KNeighborsRegressor

#Creating a KNN Classifier
k_nearest_neighbours = KNeighborsRegressor(n_neighbors=3)

#Train the model using the training sets
k_nearest_neighbours.fit(x_train, y_train)

#Predict the response for test dataset
y_pred_knn = k_nearest_neighbours.predict(x_train)
```

```python
rmse_knn = np.sqrt(metrics.mean_squared_error(y_train, y_pred_knn))
print('Root Mean Squared Error:', rmse_knn)
```

```python
#better RMSE of KNN model
```

```python
# plotting the K nearest neighbour model
plt.scatter(y_train, y_pred_knn, color = 'r',alpha=0.5)
plt.xlabel('True Values')
plt.ylabel('knn Predictions')
plt.title('K_N_N Model - KNN_Pred vs TV')

t = range(3,10,1)
u = range(3,10,1)
plt.plot(t,u, color = 'b', label='True = Predict', alpha = 0.5)
```

DECISION TREE REGRESSOR 

```python
#Importing Decision tree regressor from linear regressor

from sklearn.tree import DecisionTreeRegressor
d_c_r = sklearn.tree.DecisionTreeRegressor(max_depth=9)

#Train the model using the training sets
d_c_r.fit(x_train, y_train)

#Predict the response for test dataset
y_pred_dcr = d_c_r.predict(x_train)

rmse_dcr = np.sqrt(metrics.mean_squared_error(y_train, y_pred_dcr))
print('Root Mean Squared Error:', rmse_dcr)
```

CROSS VALIDATION

```python
#CROSS VALIDATION RANDOMLY SPLITS THE TRAINING DATA IN 10 DISTINCT "FOLDS",
#THEN IT TRAINS AND EVALUATED THE RESPECTIVE MODEL 10 TIMES
# PICKING A DIFFERENT FOLD FOR EVALUATION EVERY TIME AND TRAINING ON THE OTHER 9 FOLDS
# RESULT WILL BE AN ARRAY CONTAINING 10 EVALUATION SCORES
#(REFERENCE - HANDS ON MACHINE LEARNING - CHAPTER 1 - Page. 73)
```

Cross validation for Linear Regression 

```python
from sklearn.model_selection import cross_val_score

regular = LinearRegression()
#select 1st model for cross validation
regular.fit(x_train,y_train)

scores = cross_val_score(regular, x_train, y_train, cv=10, scoring='neg_mean_squared_error')
final_score1 = np.sqrt(-(scores))
diff1=final_score1.mean()-train_rmse

print(final_score1)
print('Mean CVS linear reg model :', final_score1.mean())
print('Standard deviation of CV score:', final_score1.std())
print('Difference between CV and RMSE score:',diff1)
```

```python
# Minor difference in RMSE values becz of cross validation  
```

Cross validation for K nearest Neighbours

```python
K_N_N = KNeighborsRegressor(n_neighbors=3)
#select 1st model for cross validation
scores = cross_val_score(K_N_N, x_train, y_train, cv=10, scoring='neg_mean_squared_error')
final_score2 = np.sqrt(-(scores))
diff2=final_score2.mean()-rmse_knn

print(final_score2)
print('Mean CVS linear reg model :', final_score2.mean())
print('Standard deviation of CV score:', final_score2.std())
print('Difference between CV and RMSE score:',diff2)
```

```python
# increse in RMSE value because of Cross validation 
```

Cross validation for Decision Tree Regressor 

```python
d_c_r = sklearn.tree.DecisionTreeRegressor(max_depth=9)
#select 1st model for cross validation

scores = cross_val_score(d_c_r, x_train, y_train, cv=10, scoring='neg_mean_squared_error')
final_score3 = np.sqrt(-(scores))
diff3=final_score3.mean()-rmse_dcr

print(final_score3)
print('Mean CVS linear reg model :', final_score3.mean())
print('Standard deviation of CV score:', final_score3.std())
print('Difference between CV and RMSE score:',diff3)
```

```python
#increment in RMSE value because of cross validation
```

```python
#Printing the values of RMSE from Cross validation in a tabular form 
```

```python
models = pd.DataFrame({
    'Model': ['Linear Regression :', 'KNN', 
              'Decision Tree Regressor'],
    'RMSE': [train_rmse, rmse_knn, rmse_dcr],
    'CV Score':[final_score1.mean(),final_score2.mean(),final_score3.mean()],
    'Diff': [diff1, diff2, diff3]})
models.sort_values(by='RMSE', ascending=False)
```
